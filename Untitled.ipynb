```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv("train.csv")
```

```python
# Use.info()
train_df.info()
```

```python
# use.describe()
train_df.describe(include='all')
```

```python
# use .value_counts()
train_df.nunique()
```

```python
# Clean for numeric visualizations
df = train_df[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']].dropna()
df['sex']=df['Sex'].map({'male': 0,'female' : 1})
```

```python
# Pairplot
sns.pairplot(df, hue="Survived")
plt.suptitle("Pairplot of Features", y=1.02)
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Age Histogram
sns.histplot(train_df['Age'].dropna(), bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
# Age vs Survival
sns.boxplot(x='Survived', y='Age', data=train_df)
plt.title("Age vs Survival")
plt.show()
```

```python
# Sex vs Survival
sns.countplot(x='Sex', hue='Survived', data=train_df)
plt.title("Sex vs Survival")
plt.show()
```

```python
# Observation: 
# Most survivors are female. 
# 1st class passengers had a higher survival rate. 
# Passengers who paid higher fares were more likely to survive.
```

```python
# summary
# - Female passengers had a much higher survival rate.
# - 1st class passengers were more likely to survive.
# - Younger passengers slightly more likely to survive.
# - High Fare is positively correlated with survival.
# - Features like Sex, Fare, and Pclass are important for prediction.
```
